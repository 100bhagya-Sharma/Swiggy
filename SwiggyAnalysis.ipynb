```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_file_path = 'swiggy.csv'
swiggy_data = pd.read_csv(csv_file_path)
```

```python
descriptive_stats = swiggy_data[['Price', 'Avg ratings', 'Total ratings', 'Delivery time']].describe()
print("Descriptive Statistics:")
print(descriptive_stats)
```

```python
avg_delivery_time_by_area = swiggy_data.groupby('Area')['Delivery time'].mean().reset_index()
print("Average Delivery Time by Area:")
print(avg_delivery_time_by_area.head())
```

```python
avg_ratings_distribution = swiggy_data['Avg ratings'].value_counts().sort_index().reset_index()
avg_ratings_distribution.columns = ['Avg ratings', 'Count']
print("Average Ratings Distribution:")
print(avg_ratings_distribution.head())
```

```python
total_ratings_distribution = swiggy_data['Total ratings'].value_counts().sort_index().reset_index()
total_ratings_distribution.columns = ['Total ratings', 'Count']
print("Total Ratings Distribution:")
print(total_ratings_distribution.head())
```

```python
total_ratings_distribution = swiggy_data['Total ratings'].value_counts().sort_index().reset_index()
total_ratings_distribution.columns = ['Total ratings', 'Count']
print("Total Ratings Distribution:")
print(total_ratings_distribution.head())
```

```python
correlation_matrix = swiggy_data[['Price', 'Avg ratings', 'Total ratings', 'Delivery time']].corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Key Metrics')
plt.show()
```

```python
sns.pairplot(swiggy_data[['Price', 'Avg ratings', 'Total ratings', 'Delivery time']])
plt.show()
```

```python
correlation_ratings_delivery_time = swiggy_data['Avg ratings'].corr(swiggy_data['Delivery time'])
print("Correlation between Avg Ratings and Delivery Time:")
print(correlation_ratings_delivery_time)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(swiggy_data['Delivery time'], kde=True, bins=30)
plt.title('Distribution of Delivery Time')
plt.xlabel('Delivery Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(swiggy_data['Avg ratings'], kde=True, bins=30)
plt.title('Distribution of Average Ratings')
plt.xlabel('Average Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(swiggy_data['Price'], kde=True, bins=30)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
avg_delivery_time_by_area = swiggy_data.groupby('Area')['Delivery time'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=avg_delivery_time_by_area, x='Delivery time', y='Area')
plt.title('Average Delivery Time by Area')
plt.xlabel('Average Delivery Time (minutes)')
plt.ylabel('Area')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
avg_ratings_by_food_type = swiggy_data.groupby('Food type')['Avg ratings'].mean().sort_values(ascending=False).reset_index()
sns.barplot(data=avg_ratings_by_food_type, x='Avg ratings', y='Food type')
plt.title('Average Ratings by Food Type')
plt.xlabel('Average Ratings')
plt.ylabel('Food Type')
plt.show()
```
